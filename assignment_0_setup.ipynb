### 3.1 Notebook 1 Summary

Python as a whole is relatively easy to read and learn, with few lines of code which make it easy to identify potential errors, while still being powerful enough for big data with a vast array of libraries. It uses 0-based indexing which makes some operations easier than if indexing started with 1. 

You can easily assign variables data using the “=” operator. After assigning a variable, you can manipulate the data through other operators like “*”, which multiplies, “+=” which is an increment operator (addition), “-=” which is a decrement operator (subtraction), and “**” for exponents. 

In python, numbers can be floats (decimals) or integers (whole numbers). You can see the type of number by inputting “type()”, and assign a variable to be integer or float by assigning x = int(3.2) or x = float(2.3). Integers have a limited range, but can be converted to long type, as python does automatically when needed. Booleans can also be used, with logical operations to specify “true” or “false”. For example, if you set x = 5, and check if x is exactly equal to 5 (x==5) it should return a “true” whereas if you check if x is exactly equal to 6 (x==6) it should return a false.

In addition to more simple operators, one can also utilize inequalities “!=”, do comparisons “<=”, or even negate with “NOT” before a line of code. 

If one wants to define a string, they can use double quotes surrounding a sentence or characters like x = “I am excited to start learning Python!”, one can use triple quotes for printing on multiple lines. One can check the length of a string “len()”, print the string “print()”, select part of a string “sentence[6:10]”, add strings “ I am excited” + “to start learning Python!”, or even convert other types to strings such as “str(3.14)”. Strings are circular, can be indexed, and are immutable. 
Container types are lists, tuples, sets and dictionaries. Lists can contain any type, such as integers, floats, strings, etc. Just like strings, these are iterable, meaning we can access certain indexes through code like “int_list[3]” which will pull the third part of the list. Lists can be nested “x = [[1,2,3],[4,5,6]]”, or added to with the append code. These are mutable, meaning you can change a certain index to a new assignemnt. Tuples are similar in that they are iterable, but they are not mutable. Sets are also similar to lists but cannot contain duplicate elements. Finally, one can bring items together by using the “zip” code. If one has two lists, they can zip them together to create tuples!
